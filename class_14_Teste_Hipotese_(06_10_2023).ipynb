# Teste de Hipótese

## Biblioteca

```python
import numpy as np
import pandas as pd
from scipy import stats as st #teste de hipotese
```

Dados

```python
pd.read_csv('salarios_profs.csv', encoding = 'latin1', sep = ';')
```

```python
dados_salarios = pd.read_csv('salarios_profs.csv',
                             encoding = 'latin1',
                             sep = ';',
                             decimal = ',')
```

```python
dados_salarios[['NOME', 'UNID_ORGAO','FUNCAO','SALARIO_MENSAL_LIQUIDO']] \
              .sort_values('SALARIO_MENSAL_LIQUIDO', ascending = False).head()
```

Teste 1


```python
dados_filtrados = dados_salarios[(dados_salarios['FUNCAO'] == 'Professor Titular') & \
               (dados_salarios['JORNADA'] == 'RDIDP') & \
               (dados_salarios['UNID_ORGAO'] == 'ICMC')]
```

```python
dados_filtrados = dados_filtrados[['UNID_ORGAO', 'SALARIO_MENSAL_LIQUIDO']]
```

```python
dados_filtrados
```

```python
dados_filtrados.groupby('UNID_ORGAO') \
                    .agg(media_salarial = pd.NamedAgg('SALARIO_MENSAL_LIQUIDO', 'mean'),
                     dp_salarial = pd.NamedAgg('SALARIO_MENSAL_LIQUIDO', 'std'),
                     n = pd.NamedAgg('SALARIO_MENSAL_LIQUIDO', 'count'))
```

```python
#teste de hipotese
#a = media  popmean, alternative = Hipotese Alternativa
# pvalue = area de rejeição a partir do t critico
st.ttest_1samp(a = dados_filtrados['SALARIO_MENSAL_LIQUIDO'],
               popmean = 20000,
               alternative = 'greater')
```

```python
# confiança:
# 90 = 0,1
# 95 = 0,05
# 99 = 0,01
# compara com o pvalue
```

Teste 2 - Duas médias

```python
dados_filtrados2 = dados_salarios[(dados_salarios['FUNCAO'] == 'Professor Titular') & \
               (dados_salarios['JORNADA'] == 'RDIDP') & \
               (dados_salarios['UNID_ORGAO'].isin(['ICMC', 'FEA']))]
```

```python
dados_filtrados2.groupby('UNID_ORGAO') \
                    .agg(media_salarial = pd.NamedAgg('SALARIO_MENSAL_LIQUIDO', 'mean'),
                     dp_salarial = pd.NamedAgg('SALARIO_MENSAL_LIQUIDO', 'std'),
                     n = pd.NamedAgg('SALARIO_MENSAL_LIQUIDO', 'count'))
```

```python
gra = dados_filtrados2[dados_filtrados2['UNID_ORGAO'] == 'ICMC']['SALARIO_MENSAL_LIQUIDO']
grb = dados_filtrados2[dados_filtrados2['UNID_ORGAO'] == 'FEA']['SALARIO_MENSAL_LIQUIDO']
st.ttest_ind(a = gra, b = grb, alternative = 'two-sided')
```
