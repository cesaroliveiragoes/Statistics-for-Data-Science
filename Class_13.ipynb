```python
import numpy as np
import pandas as pd
```

Carregar Dados


```python
dados_paises = pd.read_csv('dados_1997_2011_paises_csv.csv', sep = ';', decimal = ',', encoding='latin1')
```

Estatísticas


```python
def q1_quantile(x):
  return np.quantile(x, 0.25)
```

```python
def q3_quantile(x):
  return np.quantile(x, 0.75)
```

```python
dados_paises.groupby ('pais') \
             .agg(media_idh = pd.NamedAgg('idh','mean'),
                  mediana_idh = pd.NamedAgg('idh','median'),
                  min_idh = pd.NamedAgg('idh','min'),
                  max_idh = pd.NamedAgg('idh','max'),
                  q1_idh = pd.NamedAgg('idh',q1_quantile),
                  q3_idh = pd.NamedAgg('idh',q3_quantile),
                  dp_idh = pd.NamedAgg('idh','std'))\
              .reset_index()
```
