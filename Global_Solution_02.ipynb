```python
def read_file(filepath):
    # Função que lê o conteúdo de um arquivo.

    # Parâmetros - filepath: O caminho para o arquivo a ser lido.

    with open(filepath) as f:
        # Abre o arquivo especificado pelo caminho 'filepath'.
        str_text = f.read()
        # Lê o conteúdo do arquivo e armazena na variável 'str_text'.

    return str_text
    # Retorna o conteúdo do arquivo como uma string.
```

```python
import spacy

nlp = spacy.load('en_core_web_sm', disable=['parser', 'tagger', 'ner'])
# Carrega o modelo do Spacy para processamento de texto em inglês.

nlp.max_length = 1198623
# Define o comprimento máximo de texto que o Spacy pode processar.
```

```python
def separate_punc(doc_text):
    # Função para separar pontuações de um texto.

    # Parâmetros - doc_text: O texto a ser processado.

    tokens = nlp(doc_text)
    # Utiliza o spaCy para processar o texto e extrair os tokens.

    words = []
 # Cria uma lista vazia para armazenar as palavras sem pontuações.

    for token in tokens:
    # Itera sobre cada token no texto processado.

        if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ':
          # Verifica se o texto do token não está presente na string de pontuações.

            words.append(token.text.lower())
             # Converte o texto do token para letras minúsculas e adiciona à lista de palavras.

    return words
        # Retorna a lista de palavras sem pontuações.
```

```python
d = read_file("/content/melville-moby_dick.txt")
# Lê o conteúdo do arquivo especificado no caminho e armazena na variável 'd'.
tokens = separate_punc(d)
# Separa as pontuações do texto contido em 'd' e retorna uma lista de palavras em letras minúsculas, sem as pontuações armazenando na variável 'tokens'.
```

```python
train_len = 25+1
# Define o tamanho do treinamento como 25 palavras mais 1 (total de 26 palavras).

text_sequences = []
# Cria uma lista vazia.

for i in range(train_len, len(tokens)):
  # Itera sobre os índices de train_len até o final da lista de tokens.

    seq = tokens[i-train_len:i]
      # Extrai a sequência de palavras de tamanho train_len a partir do índice i.

    text_sequences.append(seq)
      # Adiciona a sequência à lista de text_sequences.
```

```python
from keras.preprocessing.text import Tokenizer
# Importação da classe Tokenizer para vetorização de texto e pré-processamento dentro da biblioteca do Keras.
```

```python
tokenizer = Tokenizer()
# Criação e ajuste da variável tokenizer.
tokenizer.fit_on_texts(text_sequences)
# Atualiza o tokenizador com base nas sequências de texto fornecidas.

sequences = tokenizer.texts_to_sequences(text_sequences)
# Vetorização das sequências de texto.
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding,Dropout
# Importação das classes necessárias para construir o modelo LSTM

# Sequential: Essa classe permite criar um modelo sequencial, onde as camadas são organizadas em uma sequência linear, uma após a outra.
# Dense: Essa classe representa uma camada de neurônios totalmente conectada. Ela é usada para adicionar uma camada densa, onde cada neurônio está conectado a todos os neurônios da camada anterior.
# LSTM: Essa classe representa uma camada LSTM, que é um tipo especial de camada em redes neurais recorrentes (RNNs). Ela é especialmente útil para lidar com dados sequenciais, como séries temporais ou texto.
# Embedding: Essa classe é usada para criar representações vetoriais densas para palavras ou tokens. Ela mapeia cada palavra ou token para um vetor de números reais em um espaço vetorial.
# Dropout: Essa classe implementa a técnica de dropout, que é uma forma de regularização usada para evitar overfitting durante o treinamento do modelo. O dropout aleatoriamente desativa um certo número de neurônios durante o treinamento, o que ajuda a evitar a dependência excessiva de neurônios individuais.
```

```python
def create_model(vocabulary_size, seq_len):
    #Função para criar um modelo LSTM.

    #Parâmetros - vocabulary_size: O tamanho do vocabulário, ou seja, o número total de palavras únicas no conjunto de dados  e  - seq_len: O comprimento máximo de sequência a ser considerado pelo modelo.

    model = Sequential()
    # Cria uma instância do modelo sequencial.

    model.add(Embedding(vocabulary_size, 25, input_length=seq_len))
    # Adiciona uma camada de embedding ao modelo. A camada de embedding mapeia cada palavra para um vetor de tamanho 25.

    model.add(LSTM(150, return_sequences=True))
    # Adiciona uma camada LSTM ao modelo com 150 unidades. A opção 'return_sequences=True' indica que essa camada retorna sequências em vez de apenas o último estado.

    model.add(LSTM(150))
    # Adiciona uma segunda camada LSTM ao modelo com 150 unidades.

    model.add(Dense(150, activation='relu'))
    # Adiciona uma camada densa (totalmente conectada) ao modelo com 150 unidades e função de ativação ReLU.

    model.add(Dense(vocabulary_size, activation='softmax'))
    # Adiciona uma camada densa de saída com um neurônio para cada palavra do vocabulário e função de ativação softmax.

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    # Compila o modelo com a função de perda 'categorical_crossentropy', o otimizador 'adam' e a métrica de avaliação 'accuracy'.

    model.summary()
    # Imprime um resumo do modelo, mostrando a arquitetura das camadas e o número de parâmetros.

    return model
    # Retorna o modelo LSTM criado.
```

```python
def generate_text(model, tokenizer, seq_len, seed_text, num_gen_words):
    #Função que gera texto usando um modelo treinado.

    #Parâmetros:
    - model: Modelo treinado com os dados de texto.
    - tokenizer: Tokenizer que foi ajustado com os dados de texto.
    - seq_len: Comprimento da sequência de treinamento.
    - seed_text: Texto inicial para servir como base.
    - num_gen_words: Número de palavras a serem geradas pelo modelo.

    output_text = []
    # Saída final

    input_text = seed_text
     # Sequência inicial

    for i in range(num_gen_words):
        # Criação do texto gerado

        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
         # Converte o texto de entrada em uma sequência codificada

        pad_encoded = pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')
        # Preenche as sequências de acordo com o comprimento de treinamento (seq_len)

        pred_word_ind = model.predict_classes(pad_encoded, verbose=0)[0]
        # Prediz as probabilidades de classe para cada palavra

        pred_word = tokenizer.index_word[pred_word_ind]
        # Obtém a palavra prevista

        input_text += ' ' + pred_word
        # Atualiza a sequência de texto de entrada (deslocando uma palavra com a nova palavra)

        output_text.append(pred_word)
        # Adiciona a palavra prevista 'pred_word' à lista 'output_text'

    # Retorna o texto gerado como uma única string
    return ' '.join(output_text)
```
