# 3 – Quantas compras diferentes cada cliente realizou? Quais os top 10?

```python
dados_compra = dados[['CustomerID', 'StockCode']]
dados_compra
```

```python
# Compras diferentes que cada cliente realizou

dados_compra_agrupado = dados_compra.groupby('CustomerID') \
.agg(n_compras = pd.NamedAgg('StockCode', 'nunique'))

dados_compra_agrupado
```

```python
# Top 10 compras diferentes cada cliente realizou

dados_compra_agrupado.sort_values('n_compras', ascending = False).head(10)
```

4 - Quais os top 10 clientes que mais compraram em quantidades? Quais os top 10 clientes que mais compraram em valor?

```python
# Top 10 clientes que mais compraram em quantidades

dados_qtd = dados[['CustomerID', 'Quantity']]
dados_qtd
```

```python
dados_qtd_agrupado = dados_qtd.groupby('CustomerID') \
.agg(n_qtd = pd.NamedAgg('Quantity', 'nunique'))

dados_qtd_agrupado
```

```python
dados_qtd_agrupado.sort_values('n_qtd', ascending = False).head(10)
```

```python
# Top 10 clientes que mais compraram em valor

dados_valor = dados[['CustomerID', 'UnitPrice', 'Quantity']]
dados_valor
```

```python
dados_valor['Valor'] = dados_valor['UnitPrice'] * dados_valor['Quantity']
dados_valor
```

```python
dados_valor_agrupado = dados_valor.groupby('CustomerID') \
.agg(n_valor = pd.NamedAgg('Valor', 'nunique'))

dados_valor_agrupado
```

```python
dados_valor_agrupado.sort_values('n_valor', ascending = False).head(10)
```

5 – Um dos produtos possui a seguinte função lucro −𝟐𝒑𝟐+𝟓𝟎𝒑−𝟏𝟎. Encontre a derivada da função lucro. Qual seria o melhor preço?

```python
# definir o simbolo

p = sp.Symbol('p')
p
```

```python
# definir a equação

funcao = (-2*p**2) + (50*p) - 10
funcao
```

```python
# encontrar a derivada
funcao_derivada = sp.diff(funcao)
funcao_derivada
```

```python
# simplificar a funcao

sp.simplify(funcao_derivada)
```

```python
# 50 - 4p
# -4p = -50
p = (-50) / (-4)

print(f'O melhor preço seria {p}')
```
